```python
import pandas as pd
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
movies.shape
```

```python
#We will filter the movies which have a duration of at least 200
type(True)
```

```python
#We will create a list of booleans that will be the same lenght as the column
#of duratios, with 1's for the ones with at least 200 duration
booleans = []
for length in movies.duration:
    if length >= 200:
        booleans.append(True)
    else:
        booleans.append(False)
```

```python
booleans[0:5]
```

```python
len(booleans)
```

```python
#converts booleans into a pandas Series
is_long = pd.Series(booleans)
```

```python
is_long.head()
```

```python
movies[is_long] #We are passing a series of true and falses
#result, a frame with all rows with at least 200 duration
```
